```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset=dataset.drop("User_ID",axis=1)
```

```python
indepCols=dataset[["Age","EstimatedSalary","Gender_Male"]]
depCols=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(indepCols, depCols, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':['gini','entropy'],
              'max_features': ['auto','sqrt','log2'],
              'splitter':['best','random']} 



grid = GridSearchCV(DecisionTreeClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
# fitting the model for grid search 
grid.fit(x_train, y_train) 
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(x_test) 

```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
